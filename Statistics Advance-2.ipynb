## Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

Probability Mass Function (PMF) and Probability Density Function (PDF) are both mathematical functions used to describe the probability distribution of a random variable. However, they are used for different types of random variables.

Probability Mass Function (PMF) is used to describe the probability distribution of a discrete random variable. It gives the probability that a discrete random variable takes on a specific value. The PMF is defined as:

P(X = x) = probability of X taking on the value x

For example, consider rolling a fair six-sided die. The random variable X can take on the values 1, 2, 3, 4, 5, or 6. The PMF of X is given by:

P(X = 1) = 1/6

P(X = 2) = 1/6

P(X = 3) = 1/6

P(X = 4) = 1/6

P(X = 5) = 1/6

P(X = 6) = 1/6

The sum of all the probabilities is equal to 1, since the die must land on exactly one of the six possible outcomes.

Probability Density Function (PDF) is used to describe the probability distribution of a continuous random variable. It gives the probability density at each possible value of the continuous random variable. Unlike the PMF, the PDF does not give the probability of a specific value, but rather the probability of a value falling within a specific range. The PDF is defined as:

f(x) = probability density at x

For example, consider the normal distribution with mean μ = 0 and standard deviation σ = 1. The PDF of the normal distribution is given by:

f(x) = (1/√(2πσ^2)) * e^(-((x-μ)^2)/(2σ^2))

The PDF is bell-shaped and symmetric about the mean. The total area under the curve is equal to 1, since the probability of any possible value falling between negative and positive infinity is 1.





## Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

Cumulative Density Function (CDF) is a function that gives the probability that a random variable X is less than or equal to a certain value x. In other words, it is a function that shows the probability distribution of a random variable. It is defined as the integral of the probability density function (PDF) of X from negative infinity to x.

CDF(x) = P(X <= x) = integral of PDF from negative infinity to x


The CDF is useful in statistical analysis as it can be used to calculate the probability of a random variable taking a value between two given values, or outside of a given range. It is also used to calculate other statistical measures such as percentiles and moments.


For example, let's consider a standard normal distribution with a mean of 0 and a standard deviation of 1. The CDF for this distribution can be calculated using the formula:


CDF(x) = (1 / (sqrt(2 * pi))) * integral of e^((-t^2) / 2) from negative infinity to x


If we want to find the probability of a value between -1 and 1, we can calculate the CDF at these values and subtract the two probabilities:

P(-1 < X < 1) = CDF(1) - CDF(-1)


Using a statistical software package or calculator, we can find that CDF(1) is approximately 0.8413 and CDF(-1) is approximately 0.1587. Therefore, the probability of a value between -1 and 1 in a standard normal distribution is 0.6826.

In this way, CDF provides a way to calculate probabilities and make statistical inferences based on the distribution of a random variable.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the mean and standard deviation of the normal distribution
mu = 0
sigma = 1

# Create a range of x values to evaluate the CDF
x = np.linspace(-5, 5, 1000)

# Calculate the CDF using the cumulative distribution function from scipy.stats
cdf = norm.cdf(x, mu, sigma)

# Plot the CDF
plt.plot(x, cdf)
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('Normal Distribution CDF')
plt.show()
```

## Q8: What is the z score? State the importance of the z score.

The z-score is a standardized score that indicates how many standard deviations an observation or data point is from the mean of the dataset. It is calculated by subtracting the mean of the dataset from the observation and dividing the result by the standard deviation of the dataset.

The importance of the z-score lies in the fact that it allows us to compare observations from different datasets that may have different means and standard deviations. By standardizing the observations using the z-score, we can compare how far an observation is from its respective mean relative to other observations in the dataset.

The z-score is also used in hypothesis testing, where it is used to determine whether an observation is statistically significant or not. A z-score greater than or less than a certain threshold can indicate that the observation is significantly different from the mean of the dataset, which may be evidence in support of a hypothesis or theory.

In summary, the z-score is an important statistical measure that standardizes observations and allows for meaningful comparisons between datasets and hypothesis testing.

## Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

Central Limit Theorem (CLT) is a statistical theory that describes the behavior of sample means from any distribution as the sample size increases. According to the theorem, as the sample size increases, the distribution of the sample means approaches a normal distribution, regardless of the shape of the original population distribution. This means that even if the population distribution is not normally distributed, if we take sufficiently large samples, the sample means will still follow a normal distribution.

The significance of the Central Limit Theorem is that it allows us to make statistical inferences about population parameters, such as the mean and standard deviation, based on sample means. It also provides a framework for hypothesis testing and confidence interval estimation, which are important tools in statistical analysis. In essence, the Central Limit Theorem provides a way to analyze data from a wide variety of distributions using the normal distribution, which is well understood and has many useful properties.

## Q10: State the assumptions of the Central Limit Theorem.

The Central Limit Theorem (CLT) has the following assumptions:

1. The sample size should be sufficiently large. Generally, a sample size of 30 or more is considered large enough.
2. The sample should be random and independent. Each observation should be independent of the others.
3. The sample should be collected from a population that is not too skewed or has outliers. A normal population is not required; a population with any distribution will do.

If these assumptions are met, the Central Limit Theorem states that the sample means will be approximately normally distributed, regardless of the population distribution, with a mean equal to the population mean and a standard deviation equal to the population standard deviation divided by the square root of the sample size.
